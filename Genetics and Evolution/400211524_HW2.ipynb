```python
import matplotlib.pyplot as plt
```

```python
n=0.5
r=0.8

x=[]
y=[]

for j in range (401):
    r=j/100
    n=0.5
    for i in range(500):
        n=r*(n-(n*n))
        if i>100:
            x.append(r)
            y.append(n)
        
    
plt.scatter(x, y,s=0.01,color="red")

plt.xlabel('r')
plt.ylabel('N')

plt.show()
    
```

```python
def pop_itter_plot(n,r):
    
    x=[]
    y=[]

    for i in range(60):
        x.append(i+1)
        y.append(n)
        n=r*(n-(n*n))  
    plt.plot(x, y,color="blue")
    plt.xlabel('itteration')
    plt.ylabel('population')
    return plt
```

```python
n=0.5
r=0.8
p=pop_itter_plot(n,r)
p.scatter([30],[0],color="red")
p.show()
    
```

```python
n=0.4
r=1.5
p=pop_itter_plot(n,r)
p.scatter([30],[(r-1)/r],color="red")
p.show()
```
